Attraverso syntax comprehension si può scrivere ciò che viene scritto in più linee con una sola

```python
'''Dato il programma scritto nel seguente modo
quadrati = []
for n in range(11):
    quadrati.append(n ** 2)
    Possiamo scrivere nello stesso modo:'''
quadrati = [n ** 2 for n in range(11)]
print(quadrati)
```

*zip()* mette automaticamente due elementi per posizione in un dizionario

```python
d = dict((k,v) for k,v in zip(range(5),'abcde'))
print(d)
```

Si possono mettere delle condizioni all'interno delle syntax comprehension

```python
l = [x for x in range(1, 19) if x % 2 == 0]
print(l)
```

Le classi python si creano nel seguente modo:

```python
class C1:
    a = 8
    b = 'pulcino'
    def stampa_a(self):
        a = 10
        print(self.a) # stampo a all'interno della classe
        print(a) #stampo l'a della funzione

c1 = C1()
c1.stampa_a()
```

Il costruttore in python è *\_\_init\_\_*

```python
class C1:
    a = 8

    def __init__(self):
        self.a = 12
    def stampa_a(self):
        a = 10
        print(self.a) # stampo a all'interno della classe
        print(a) #stampo l'a della funzione

c1 = C1()
c1.stampa_a()
```
